```python
# ============================================================
# 🚀 JOB SALARY ANALYSIS — OPENROUTER VERSION
# Tool 1: Filter → Tool 2: Graphs → Tool 3: LLM → Final Para
# ============================================================

# ─────────────────────────────────────────────────────────────
# STEP 0: install
# ─────────────────────────────────────────────────────────────
# !pip install openai pandas matplotlib seaborn -q

# ─────────────────────────────────────────────────────────────
# STEP 1: CSV Upload
# ─────────────────────────────────────────────────────────────
# METHOD A — Local se direct upload:
# from google.colab import files
# files.upload()

# METHOD B — Google Drive se:
# from google.colab import drive
# drive.mount('/content/drive')
# CSV_FILE = '/content/drive/My Drive/job_salary_prediction_dataset.csv'

# ─────────────────────────────────────────────────────────────
# IMPORTS
# ─────────────────────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from openai import OpenAI   # OpenRouter uses OpenAI-compatible API
import warnings
warnings.filterwarnings('ignore')

# Dark theme
plt.rcParams['figure.facecolor'] = '#0f0f1a'
plt.rcParams['axes.facecolor']   = '#1a1a2e'
plt.rcParams['axes.edgecolor']   = '#444466'
plt.rcParams['text.color']       = '#e0e0ff'
plt.rcParams['axes.labelcolor']  = '#e0e0ff'
plt.rcParams['xtick.color']      = '#aaaacc'
plt.rcParams['ytick.color']      = '#aaaacc'
plt.rcParams['axes.titlecolor']  = '#ffffff'

print("✅ Libraries loaded!")

# ─────────────────────────────────────────────────────────────
# OPENROUTER CLIENT SETUP
# ─────────────────────────────────────────────────────────────
# OpenRouter is compatible API of OpenAi

def get_openrouter_client(api_key):

    client = OpenAI(
        base_url="https://openrouter.ai/api/v1",
        api_key=api_key,
    )
    return client

def call_llm(client, prompt, model="meta-llama/llama-3.3-8b-instruct:free", max_tokens=700):
    """

    Free models available on OpenRouter:
    - "meta-llama/llama-3.3-8b-instruct:free"    (recommended)
    - "google/gemma-3-4b-it:free"
    - "mistralai/mistral-7b-instruct:free"
    - "deepseek/deepseek-r1:free"

    Paid models:
    - "anthropic/claude-sonnet-4-5"
    - "openai/gpt-4o-mini"
    - "google/gemini-flash-1.5"
    """
    response = client.chat.completions.create(
        model=model,
        max_tokens=max_tokens,
        messages=[{"role": "user", "content": prompt}],
        extra_headers={
            "HTTP-Referer": "https://colab.research.google.com",
            "X-Title": "Job Salary Analysis"
        }
    )
    return response.choices[0].message.content


# ╔══════════════════════════════════════════════════════════╗
# ║         TOOL 1: DATA FILTERING                          ║
# ╚══════════════════════════════════════════════════════════╝

def tool_data_filter(filepath):
    """
    TOOL 1 — DATA FILTER
    - CSV load
    - Nulls and duplicates remove
    - Salary outliers remove (1%-99%)
    - Key stats remove
    """
    print("\n" + "═"*55)
    print("🔧 TOOL 1: DATA FILTERING...")
    print("═"*55)

    df = pd.read_csv(filepath)
    original_count = len(df)
    print(f"   📂 Raw rows          : {original_count:,}")

    df.drop_duplicates(inplace=True)
    print(f"   🗑️  Duplicates removed: {original_count - len(df):,}")

    before_null = len(df)
    df.dropna(inplace=True)
    print(f"   🧹 Nulls removed     : {before_null - len(df):,}")

    q_low  = df['salary'].quantile(0.01)
    q_high = df['salary'].quantile(0.99)
    before = len(df)
    df = df[(df['salary'] >= q_low) & (df['salary'] <= q_high)]
    print(f"   📉 Outliers removed  : {before - len(df):,}")
    print(f"   ✅ Clean rows        : {len(df):,}")

    summary = {
        "total_rows"    : len(df),
        "avg_salary"    : round(df['salary'].mean(), 2),
        "median_salary" : round(df['salary'].median(), 2),
        "max_salary"    : int(df['salary'].max()),
        "min_salary"    : int(df['salary'].min()),
        "top_job"       : df.groupby('job_title')['salary'].mean().idxmax(),
        "top_industry"  : df.groupby('industry')['salary'].mean().idxmax(),
        "avg_experience": round(df['experience_years'].mean(), 1),
        "edu_counts"    : df['education_level'].value_counts().to_dict(),
        "remote_counts" : df['remote_work'].value_counts().to_dict(),
        "company_size"  : df['company_size'].value_counts().to_dict(),
    }

    print(f"\n   💰 Avg Salary   : ${summary['avg_salary']:,.0f}")
    print(f"   🏆 Top Job      : {summary['top_job']}")
    print(f"   🏭 Top Industry : {summary['top_industry']}")
    print("\n✅ TOOL 1 COMPLETE!\n")
    return df, summary


# ╔══════════════════════════════════════════════════════════╗
# ║         TOOL 2: INSIGHTS GENERATOR (4 GRAPHS)          ║
# ╚══════════════════════════════════════════════════════════╝

def tool_insights_generator(df, summary):
    """
    TOOL 2 — GRAPHS BANAO
    Graph 1: Top 10 Jobs by Avg Salary
    Graph 2: Salary by Education (Box Plot)
    Graph 3: Experience vs Salary (Scatter)
    Graph 4: Remote Work + Company Size impact
    """
    print("═"*55)
    print("📊 TOOL 2: INSIGHTS GENERATOR...")
    print("═"*55)

    fig = plt.figure(figsize=(20, 16))
    fig.patch.set_facecolor('#0f0f1a')
    fig.suptitle('💼 JOB SALARY PREDICTION — DASHBOARD',
                 fontsize=20, fontweight='bold', color='#ffffff', y=0.98)

    COLORS = ['#7c3aed','#6d28d9','#5b21b6','#4c1d95',
              '#312e81','#1e40af','#1d4ed8','#2563eb','#3b82f6','#60a5fa']

    # ── Graph 1: Top 10 Jobs ──────────────────────────────
    ax1 = fig.add_subplot(2, 2, 1)
    top_jobs = (df.groupby('job_title')['salary']
                  .mean().sort_values(ascending=False).head(10))
    bars = ax1.barh(top_jobs.index, top_jobs.values,
                    color=COLORS, edgecolor='#333355', height=0.7)
    for bar, val in zip(bars, top_jobs.values):
        ax1.text(val + 300, bar.get_y() + bar.get_height()/2,
                 f'${val:,.0f}', va='center', fontsize=8,
                 color='#e0e0ff', fontweight='bold')
    ax1.set_title('Top 10 Jobs by Avg Salary', fontsize=12, fontweight='bold')
    ax1.invert_yaxis()
    ax1.xaxis.set_major_formatter(plt.FuncFormatter(lambda x,_: f'${x/1000:.0f}K'))
    ax1.grid(axis='x', alpha=0.2, linestyle='--')
    print("   ✅ Graph 1: Top Jobs")

    # ── Graph 2: Education Box Plot ───────────────────────
    ax2 = fig.add_subplot(2, 2, 2)
    edu_order = [e for e in ['High School','Bachelor','Master','PhD']
                 if e in df['education_level'].unique()]
    data_edu  = [df[df['education_level']==e]['salary'].values for e in edu_order]
    edu_colors = ['#6366f1','#8b5cf6','#a855f7','#ec4899']
    bp = ax2.boxplot(data_edu, labels=edu_order, patch_artist=True, notch=True,
                     medianprops=dict(color='white', linewidth=2))
    for patch, color in zip(bp['boxes'], edu_colors):
        patch.set_facecolor(color); patch.set_alpha(0.75)
    for el in ['whiskers','caps','fliers']:
        for item in bp[el]: item.set(color='#aaaacc', linewidth=1)
    ax2.set_title('Salary by Education Level', fontsize=12, fontweight='bold')
    ax2.set_ylabel('Salary ($)')
    ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x,_: f'${x/1000:.0f}K'))
    ax2.grid(axis='y', alpha=0.2, linestyle='--')
    print("   ✅ Graph 2: Education")

    # ── Graph 3: Experience vs Salary Scatter ─────────────
    ax3 = fig.add_subplot(2, 2, 3)
    sample = df.sample(min(3000, len(df)), random_state=42)
    edu_map = {'High School':'#6366f1','Bachelor':'#8b5cf6',
               'Master':'#a855f7','PhD':'#ec4899'}
    for edu, grp in sample.groupby('education_level'):
        ax3.scatter(grp['experience_years'], grp['salary'],
                    c=edu_map.get(edu,'#888888'), alpha=0.3, s=12, label=edu)
    z = np.polyfit(sample['experience_years'], sample['salary'], 1)
    p = np.poly1d(z)
    xs = np.linspace(sample['experience_years'].min(),
                     sample['experience_years'].max(), 100)
    ax3.plot(xs, p(xs), color='#f97316', linewidth=2.5,
             linestyle='--', label='Trend')
    ax3.set_title('Experience vs Salary', fontsize=12, fontweight='bold')
    ax3.set_xlabel('Years of Experience')
    ax3.set_ylabel('Salary ($)')
    ax3.yaxis.set_major_formatter(plt.FuncFormatter(lambda x,_: f'${x/1000:.0f}K'))
    ax3.legend(fontsize=8, framealpha=0.3, loc='upper left')
    ax3.grid(alpha=0.15, linestyle='--')
    print("   ✅ Graph 3: Experience Scatter")

    # ── Graph 4: Remote Work + Company Size ───────────────
    ax4 = fig.add_subplot(2, 2, 4)
    pivot = (df.groupby(['remote_work','company_size'])['salary']
               .mean().unstack(fill_value=0))
    x_labels = pivot.index.tolist()
    x        = np.arange(len(x_labels))
    bar_cols  = pivot.columns.tolist()
    width     = 0.25
    palette   = ['#6366f1','#f97316','#22c55e','#ec4899']
    for i, (col, color) in enumerate(zip(bar_cols, palette)):
        offset = (i - len(bar_cols)/2) * width + width/2
        ax4.bar(x + offset, pivot[col], width=width,
                label=col, color=color, alpha=0.85, edgecolor='#333355')
    ax4.set_title('Remote Work & Company Size Impact', fontsize=12, fontweight='bold')
    ax4.set_xlabel('Remote Work Type')
    ax4.set_ylabel('Avg Salary ($)')
    ax4.set_xticks(x); ax4.set_xticklabels(x_labels)
    ax4.yaxis.set_major_formatter(plt.FuncFormatter(lambda x,_: f'${x/1000:.0f}K'))
    ax4.legend(title='Company Size', fontsize=8, framealpha=0.3)
    ax4.grid(axis='y', alpha=0.2, linestyle='--')
    print("   ✅ Graph 4: Remote Work")

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.savefig('salary_dashboard.png', dpi=150,
                bbox_inches='tight', facecolor='#0f0f1a')
    plt.show()
    print("   💾 Saved: salary_dashboard.png")

    # Insights text build karo Tool 3 ke liye
    top_jobs_str = "\n".join([f"  {j}: ${s:,.0f}"
                               for j,s in top_jobs.head(5).items()])
    edu_avg    = df.groupby('education_level')['salary'].mean().round(0)
    remote_avg = df.groupby('remote_work')['salary'].mean().round(0)

    insights_text = f"""
DATASET: {summary['total_rows']:,} records after cleaning
Average Salary : ${summary['avg_salary']:,}
Median Salary  : ${summary['median_salary']:,}
Avg Experience : {summary['avg_experience']} years

TOP 5 PAYING JOBS:
{top_jobs_str}

SALARY BY EDUCATION:
{edu_avg.to_string()}

SALARY BY REMOTE TYPE:
{remote_avg.to_string()}

TOP INDUSTRY: {summary['top_industry']}
COMPANY SIZES: {summary['company_size']}
"""
    print("\n✅ TOOL 2 COMPLETE!\n")
    return insights_text


# ╔══════════════════════════════════════════════════════════╗
# ║         TOOL 3: LLM ANALYZER (BULLET POINTS)          ║
# ╚══════════════════════════════════════════════════════════╝

def tool_llm_analyzer(client, insights_text, model):
    """
    TOOL 3 — LLM BULLET ANALYSIS
    """
    print("═"*55)
    print("🤖 TOOL 3: LLM ANALYZER (OpenRouter)...")
    print("═"*55)
    print("   📡 OpenRouter is connecting...")

    prompt = f"""You are a data analyst. Analyze this Job Salary dataset summary:

{insights_text}

Give EXACTLY 6 bullet points. Each must:
- Start with a relevant emoji
- Be 1-2 sentences, clear and insightful
- Focus on salary trends, key patterns, actionable insights

Format:
• [emoji] [insight]
(6 total, no extra text before or after)"""

    bullets = call_llm(client, prompt, model=model, max_tokens=600)

    print("\n   📋 BULLET ANALYSIS:")
    print("   " + "─"*48)
    for line in bullets.strip().split('\n'):
        if line.strip():
            print(f"   {line}")
    print("   " + "─"*48)
    print("\n✅ TOOL 3 COMPLETE!\n")
    return bullets


# ╔══════════════════════════════════════════════════════════╗
# ║         FINAL: PARAGRAPH GENERATOR (~100 words)        ║
# ╚══════════════════════════════════════════════════════════╝

def generate_final_paragraph(client, bullet_analysis, summary, model):
    """
    FINAL STEP — Bullets convert ~100 word professional paragraph
    """
    print("═"*55)
    print("📝 FINAL: PARAGRAPH GENERATION...")
    print("═"*55)

    prompt = f"""Based on these job salary insights:

{bullet_analysis}

Key numbers: avg salary ${summary['avg_salary']:,},
top industry {summary['top_industry']},
avg experience {summary['avg_experience']} yrs.

Write a professional summary in EXACTLY 100 words.
- Flowing prose only, no bullet points
- Cover: salary range, education impact, experience trend, top jobs
- End with a forward-looking sentence
- No preamble, just the paragraph"""

    paragraph = call_llm(client, prompt, model=model, max_tokens=250)

    print("\n" + "═"*55)
    print("🎯 FINAL RESULT — 100 WORD PARAGRAPH:")
    print("═"*55)
    print(f"\n{paragraph}\n")
    print("═"*55)
    print("✅ PIPELINE COMPLETE!\n")
    return paragraph


# ╔══════════════════════════════════════════════════════════╗
# ║              MAIN PIPELINE — RUN KARO                  ║
# ╚══════════════════════════════════════════════════════════╝

def run_pipeline(csv_path, openrouter_api_key,
                 model="meta-llama/llama-3.3-8b-instruct:free"):
    """
    all tools in chain:
    Filter → Graphs → LLM Bullets → Final Paragraph

    """
    print("\n" + "🚀"*27)
    print("  JOB SALARY PIPELINE — OPENROUTER VERSION")
    print("🚀"*27 + "\n")
    print(f"   Model: {model}\n")

    # Client banao
    client = get_openrouter_client(openrouter_api_key)

    # Chain Step 1
    df_clean, summary = tool_data_filter(csv_path)

    # Chain Step 2
    insights_text = tool_insights_generator(df_clean, summary)

    # Chain Step 3
    bullet_analysis = tool_llm_analyzer(client, insights_text, model)

    # Final
    final_para = generate_final_paragraph(client, bullet_analysis, summary, model)

    print("\n📦 PIPELINE SUMMARY:")
    print(f"   Tool 1 → {summary['total_rows']:,} rows filtered")
    print(f"   Tool 2 → 4 graphs generated")
    print(f"   Tool 3 → 6 bullet points via {model}")
    print(f"   Final  → ~100 word paragraph")
    print("\n🎉 ALL DONE!")

    return {
        "clean_data"     : df_clean,
        "summary"        : summary,
        "insights_text"  : insights_text,
        "bullet_analysis": bullet_analysis,
        "final_paragraph": final_para
    }


# ╔══════════════════════════════════════════════════════════╗
# ║              ▶️  YAHAN SE CHALAO                       ║
# ╚══════════════════════════════════════════════════════════╝

MY_OPENROUTER_KEY = "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"

# ⚠️  CSV file ka path (upload karne ke baad)
CSV_FILE = "/content/drive/MyDrive/job_salary_prediction_dataset.csv"

# ⚠️  Model choose (FREE models )
# Free:  "meta-llama/llama-3.3-8b-instruct:free"
# Free:  "google/gemma-3-4b-it:free"
# Free:  "mistralai/mistral-7b-instruct:free"
# Paid:  "anthropic/claude-sonnet-4-5"
# Paid:  "openai/gpt-4o-mini"
MY_MODEL = "google/gemma-3-4b-it:free"

# PIPELINE RUN 🚀
results = run_pipeline(CSV_FILE, MY_OPENROUTER_KEY, MY_MODEL)
```
